Building The Model

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = [] #empty list to store the within-cluster sum of squares (WCSS) for different cluster counts
k_range = range(1, 11) #testing cluster counts from 1 to 10
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(ronin_scaled_data)
    wcss.append(kmeans.inertia_)  # Inertia: Sum of squared distances to closest cluster center

plt.figure(figsize=(10, 6))
plt.plot(list(k_range), wcss, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans.fit(ronin_scaled_data)
ronin_scaled_data['Cluster'] = kmeans.labels_
ronin_scaled_data.head()
```

```python
#calculating the silhouette score to evaluate clustering quality
from sklearn.metrics import silhouette_score

score= silhouette_score(ronin_scaled_data.drop('Cluster', axis=1), ronin_scaled_data['Cluster'])
print('Silhouette Score:', score)
```

```python
from sklearn.metrics import silhouette_score

# Empty list for silhouette scores
silhouette_scores = []

# Loop through k values from 2 to 10
k_range = range(2, 11)

for k in k_range:
    # Create and fit KMeans
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(ronin_scaled_data)
    
    # Calculate silhouette score
    score = silhouette_score(ronin_scaled_data, cluster_labels)
    silhouette_scores.append(score)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Find optimal eps value
# Rule of thumb: min_samples = 2 * number_of_features
n_features = ronin_scaled_data.shape[1]
min_samples = 2 * n_features

print(f"Number of features: {n_features}")
print(f"Recommended min_samples: {min_samples}")

# Calculate k-nearest neighbors distances
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(ronin_scaled_data)
distances, indices = neighbors_fit.kneighbors(ronin_scaled_data)

distances = np.sort(distances[:, -1], axis=0)

# Plot to find the "elbow" - this helps choose eps
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.xlabel("Points sorted by distance")
plt.ylabel(f"{min_samples}-th Nearest Neighbor Distance")
plt.title("K-Distance Graph for Optimal eps Selection")
plt.axhline(y=1.0, color='r', linestyle='--', label='Potential eps=1.0')
plt.axhline(y=1.5, color='g', linestyle='--', label='Potential eps=1.5')
plt.axhline(y=2.0, color='b', linestyle='--', label='Potential eps=2.0')
plt.legend()
plt.grid(True)
plt.show()

print("\nLook for the 'elbow' in the graph - where it sharply rises")
print("That point suggests a good eps value")
```

```python
# Trying different eps values to find better clustering
# Final DBSCAN with eps=2.5
eps_value = 2.5

dbscan_model = DBSCAN(eps=eps_value, min_samples=20, metric='euclidean')
dbscan_labels = dbscan_model.fit_predict(ronin_scaled_data)

# Add labels to dataframe
ronin_clusters_clean['dbscan_cluster'] = dbscan_labels

print(f"DBSCAN Results (eps={eps_value}, min_samples=20):")
print(f"Unique labels: {np.unique(dbscan_labels)}")
print(f"Number of clusters: {len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)}")
print(f"Number of outliers: {list(dbscan_labels).count(-1)}")

# Profile each cluster
print("\n" + "="*70)
print("DBSCAN CLUSTER PROFILES")
print("="*70)

for cluster_id in sorted(ronin_clusters_clean['dbscan_cluster'].unique()):
    cluster_data = ronin_clusters_clean[ronin_clusters_clean['dbscan_cluster'] == cluster_id]
    
    if cluster_id == -1:
        print(f"\n{'='*70}")
        print(f"OUTLIERS (Noise) - Size: {len(cluster_data)} wallets ({len(cluster_data)/len(ronin_clusters_clean)*100:.2f}%)")
        print(f"{'='*70}")
    else:
        print(f"\n{'='*70}")
        print(f"CLUSTER {cluster_id} - Size: {len(cluster_data)} wallets ({len(cluster_data)/len(ronin_clusters_clean)*100:.2f}%)")
        print(f"{'='*70}")
    
    print(f"  Transaction Count: mean={cluster_data['transaction_count'].mean():.0f}, median={cluster_data['transaction_count'].median():.0f}")
    print(f"  Days Active: mean={cluster_data['days_active'].mean():.1f}, median={cluster_data['days_active'].median():.1f}")
    print(f"  Value Sent (Total RON): mean={cluster_data['value_sent'].mean():.2f}, median={cluster_data['value_sent'].median():.2f}")
    print(f"  Average Value per Transaction: mean={cluster_data['average_value_sent'].mean():.2f}, median={cluster_data['average_value_sent'].median():.2f}")
    print(f"  Unique Addresses: mean={cluster_data['unique_address'].mean():.0f}, median={cluster_data['unique_address'].median():.0f}")
```

```python
# BUSINESS PROBLEM SOLVER, Using Clustering for Actionable Insights

import pandas as pd
import numpy as np

print("="*80)
print("RONIN WALLET CLUSTERING - BUSINESS INSIGHTS & SOLUTIONS")
print("="*80)

# We'll use K-Means as primary (best performance) and DBSCAN for anomalies
# Let's create a comprehensive business analysis


# PROBLEM 1, DETECTING POTENTIAL CHURNERS

print("\n" + "="*80)
print("PROBLEM 1: CHURN RISK DETECTION")
print("="*80)

# Define churn risk based on days_since_last_transaction
# High risk: >7 days inactive
# Medium risk: 3-7 days inactive  
# Low risk: <3 days inactive

ronin_clusters_clean['churn_risk'] = pd.cut(
    ronin_clusters_clean['days_since_last_transactions'],
    bins=[-1, 3, 7, float('inf')],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

# Analyze churn risk by cluster (using K-Means clusters)
print("\nChurn Risk by User Segment (K-Means Clusters):")
churn_analysis = ronin_clusters_clean.groupby(['cluster', 'churn_risk']).size().unstack(fill_value=0)
churn_pct = churn_analysis.div(churn_analysis.sum(axis=1), axis=0) * 100

print("\nChurn Risk Distribution (%):")
print(churn_pct.round(2))

# Identify high-risk segments
print("\n HIGH-RISK CHURN SEGMENTS:")
for cluster in churn_pct.index:
    high_risk_pct = churn_pct.loc[cluster, 'High Risk']
    cluster_size = ronin_clusters_clean[ronin_clusters_clean['cluster'] == cluster].shape[0]
    
    if high_risk_pct > 30:  # More than 30% high risk
        print(f"\n  Cluster {cluster}: {high_risk_pct:.1f}% high churn risk ({cluster_size} wallets)")
        cluster_data = ronin_clusters_clean[ronin_clusters_clean['cluster'] == cluster]
        print(f"    - Avg days inactive: {cluster_data['days_since_last_transaction'].mean():.1f}")
        print(f"    - Avg transaction count: {cluster_data['transaction_count'].mean():.0f}")
        print(f"    - Avg value sent: {cluster_data['value_sent'].mean():.2f} RON")


# PROBLEM 2, WHALE VS CASUAL USER SEGMENTS

print("\n" + "="*80)
print("PROBLEM 2: WHALE VS CASUAL USER IDENTIFICATION")
print("="*80)

# Define user tiers based on value_sent
ronin_clusters_clean['user_tier'] = pd.cut(
    ronin_clusters_clean['value_sent'],
    bins=[-1, 100, 1000, 5000, float('inf')],
    labels=['Casual', 'Regular', 'Power User', 'Whale']
)

# Analyze user tiers by cluster
print("\nUser Tier Distribution by Cluster:")
tier_analysis = ronin_clusters_clean.groupby(['cluster', 'user_tier']).size().unstack(fill_value=0)
tier_pct = tier_analysis.div(tier_analysis.sum(axis=1), axis=0) * 100

print("\nUser Tier Distribution (%):")
print(tier_pct.round(2))

print("\nðŸ‹ WHALE SEGMENTS (High-Value Users):")
for cluster in tier_pct.index:
    if 'Whale' in tier_pct.columns:
        whale_pct = tier_pct.loc[cluster, 'Whale']
        if whale_pct > 10:  # More than 10% whales
            cluster_data = ronin_clusters_clean[ronin_clusters_clean['cluster'] == cluster]
            whale_count = (cluster_data['user_tier'] == 'Whale').sum()
            total_whale_value = cluster_data[cluster_data['user_tier'] == 'Whale']['value_sent'].sum()
            
            print(f"\n  Cluster {cluster}: {whale_pct:.1f}% whales ({whale_count} wallets)")
            print(f"    - Total whale value: {total_whale_value:,.2f} RON")
            print(f"    - Avg whale value: {total_whale_value/whale_count:,.2f} RON")

print("\n CASUAL USER SEGMENTS:")
for cluster in tier_pct.index:
    casual_pct = tier_pct.loc[cluster, 'Casual']
    if casual_pct > 70:  # More than 70% casual
        cluster_data = ronin_clusters_clean[ronin_clusters_clean['cluster'] == cluster]
        casual_count = (cluster_data['user_tier'] == 'Casual').sum()
        
        print(f"\n  Cluster {cluster}: {casual_pct:.1f}% casual users ({casual_count} wallets)")
        print(f"    - Avg transactions: {cluster_data['transaction_count'].mean():.0f}")
        print(f"    - Avg days active: {cluster_data['days_active'].mean():.1f}")

# PROBLEM 3, GAMER BEHAVIOR TYPES

print("\n" + "="*80)
print("PROBLEM 3: GAMER BEHAVIOR SEGMENTATION")
print("="*80)

# Create behavior profiles for each cluster
print("\nBehavior Profile by Cluster:\n")

for cluster_id in sorted(ronin_clusters_clean['cluster'].unique()):
    cluster_data = ronin_clusters_clean[ronin_clusters_clean['cluster'] == cluster_id]
    
    # Calculate key behavioral metrics
    avg_txn = cluster_data['transaction_count'].median()
    avg_value = cluster_data['value_sent'].median()
    avg_days_active = cluster_data['days_active'].median()
    avg_unique_addr = cluster_data['unique_address'].median()
    avg_value_per_txn = cluster_data['average_value_sent'].median()
    
    # Classify behavior type
    if avg_txn > 1000 and avg_value_per_txn < 0.1:
        behavior_type = "ðŸ¤– BOT/GRINDER"
        description = "High-frequency, low-value transactions. Likely automated farming."
    elif avg_value > 5000:
        behavior_type = "ðŸ‹ WHALE/INVESTOR"
        description = "High-value transactions. Investment-focused users."
    elif avg_unique_addr > 100:
        behavior_type = "ðŸŒ EXCHANGE/HUB"
        description = "Connected to many addresses. Likely exchange or marketplace."
    elif avg_txn > 200 and avg_value > 500:
        behavior_type = "ACTIVE TRADER"
        description = "Regular trading activity with moderate-to-high values."
    else:
        behavior_type = "ðŸ’¤ CASUAL GAMER"
        description = "Typical gaming activity with low monetary value."
    
    print(f"Cluster {cluster_id}: {behavior_type}")
    print(f"  Size: {len(cluster_data)} wallets ({len(cluster_data)/len(ronin_clusters_clean)*100:.1f}%)")
    print(f"  Description: {description}")
    print(f"  Profile: {avg_txn:.0f} txns, {avg_value:.2f} RON, {avg_days_active:.0f} days active")
    print()

# PROBLEM 4, GAME DEPENDENCY ANALYSIS

print("\n" + "="*80)
print("PROBLEM 4: PLATFORM DEPENDENCY & DIVERSIFICATION")
print("="*80)

# Analyze engagement diversity (using unique_address as proxy for platform diversity)
ronin_clusters_clean['engagement_diversity'] = pd.cut(
    ronin_clusters_clean['unique_address'],
    bins=[-1, 5, 15, 50, float('inf')],
    labels=['Single Game', 'Low Diversity', 'Moderate Diversity', 'High Diversity']
)

diversity_analysis = ronin_clusters_clean.groupby(['cluster', 'engagement_diversity']).size().unstack(fill_value=0)
diversity_pct = diversity_analysis.div(diversity_analysis.sum(axis=1), axis=0) * 100

print("\nEngagement Diversity by Cluster (%):")
print(diversity_pct.round(2))

print("\nâš ï¸ HIGH DEPENDENCY RISK SEGMENTS:")
for cluster in diversity_pct.index:
    single_game_pct = diversity_pct.loc[cluster, 'Single Game']
    if single_game_pct > 50:  # More than 50% single-game users
        cluster_data = ronin_clusters_clean[ronin_clusters_clean['cluster'] == cluster]
        at_risk = (cluster_data['engagement_diversity'] == 'Single Game').sum()
        
        print(f"\n  Cluster {cluster}: {single_game_pct:.1f}% single-game dependent ({at_risk} wallets)")
        print(f"    - Risk: If main game fails, these users likely churn")
        print(f"    - Avg unique addresses: {cluster_data['unique_address'].mean():.1f}")
        print(f"    - Recommendation: Cross-promote other games to this segment")

# ACTIONABLE RECOMMENDATIONS

print("\n" + "="*80)
print("ðŸ’¡ ACTIONABLE BUSINESS RECOMMENDATIONS")
print("="*80)

print("""
Based on your clustering analysis, here are specific actions:

1. CHURN PREVENTION:
   âœ… Target high-risk clusters with re-engagement campaigns
   âœ… Send notifications to users inactive >7 days
   âœ… Offer incentives (tokens, NFTs) to at-risk whales
   âœ… A/B test different retention messages per cluster

2. WHALE RETENTION & GROWTH:
   âœ… Identify whale clusters and assign VIP support
   âœ… Create exclusive events/tournaments for high-value users
   âœ… Track whale churn in real-time dashboards
   âœ… Analyze what makes whales different from regular users

3. GAMER ENGAGEMENT:
   âœ… Customize game recommendations per behavior type
   âœ… Create progression paths for casual â†’ power users
   âœ… Design bot-resistant game mechanics (if bots detected)
   âœ… Build community features for highly engaged clusters

4. PLATFORM DIVERSIFICATION:
   âœ… Cross-promote games to single-game dependent users
   âœ… Create multi-game rewards/loyalty programs
   âœ… Bundle game purchases or subscriptions
   âœ… Track adoption of new games by cluster
   âœ… Reduce risk by expanding user interests across portfolio
""")

print("\n" + "="*80)
print("âœ… ANALYSIS COMPLETE")
print("="*80)
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import joblib
import os
from pathlib import Path

# Page configuration
st.set_page_config(
    page_title="Ronin Wallet Clustering Dashboard",
    page_icon="ðŸŽ®",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Simple CSS
st.markdown("""
<style>
    .main-header { font-size: 2.2rem; font-weight: bold; color: #1f77b4; text-align: center; }
</style>
""", unsafe_allow_html=True)

@st.cache_resource
def load_models():
    base = Path.cwd()

    def try_load(p):
        p = Path(p)
        if not p.exists():
            return None
        try:
            return joblib.load(p)
        except Exception as e:
            st.warning(f"Failed to load {p.name}: {e}")
            return None

    def is_estimator(obj):
        return hasattr(obj, 'predict') or hasattr(obj, 'fit_predict')

    saved_label_arrays = {}

    # Helper to try multiple paths without triggering truth-value checks on arrays
    def try_load_any(paths):
        for p in paths:
            res = try_load(p)
            if res is not None:
                return res
        return None

    kmeans = try_load_any([base / 'saved_models' / 'ronin_kmeans_model.pkl', base / 'models' / 'kmeans_model.pkl'])
    if kmeans is not None and not is_estimator(kmeans):
        saved_label_arrays['kmeans'] = kmeans
        kmeans = None

    agg = try_load_any([base / 'saved_models' / 'ronin_agg_model.pkl', base / 'models' / 'agg_model.pkl'])
    if agg is not None and not is_estimator(agg):
        saved_label_arrays['agg'] = agg
        agg = None

    dbscan = try_load_any([base / 'saved_models' / 'ronin_dbscan_model.pkl', base / 'models' / 'dbscan_model.pkl'])
    if dbscan is not None and not is_estimator(dbscan):
        saved_label_arrays['dbscan'] = dbscan
        dbscan = None

    mb_kmeans = try_load_any([base / 'saved_models' / 'ronin_mb_kmeans_model.pkl', base / 'models' / 'mb_kmeans_model.pkl'])
    if mb_kmeans is not None and not is_estimator(mb_kmeans):
        saved_label_arrays['mb_kmeans'] = mb_kmeans
        mb_kmeans = None

    scaler = try_load_any([base / 'saved_scalers' / 'ronin_scaler.pkl', base / 'models' / 'scaler.pkl'])
    feature_names = try_load_any([base / 'saved_scalers' / 'ronin_scaled_features.pkl', base / 'models' / 'feature_names.pkl'])

    # Agglomerative NN mapping
    agg_nn = try_load_any([base / 'saved_models' / 'agg_nn.pkl', base / 'models' / 'agg_nn.pkl'])
    agg_labels = try_load_any([base / 'saved_models' / 'agg_labels.pkl', base / 'saved_models' / 'ronin_agg_labels.pkl'])
    if agg_labels is not None:
        saved_label_arrays['agg'] = agg_labels

    return kmeans, agg, dbscan, mb_kmeans, scaler, feature_names, saved_label_arrays, agg_nn

# Use in-memory dataset when available, else try CSV
try:
    data = ronin_clusters_clean
except Exception:
    possible = [Path('ronin_clusters_clean.csv'), Path('data/ronin_clusters_clean.csv')]
    data = None
    for p in possible:
        if p.exists():
            data = pd.read_csv(p)
            break
    if data is None:
        st.error("Data not found in notebook environment. Please run training cells or place 'ronin_clusters_clean.csv' in the folder.")

kmeans_model, agg_model, dbscan_model, mb_kmeans_model, scaler, feature_names, saved_label_arrays, agg_nn = load_models()

# Sidebar status
with st.sidebar.expander("Model & Data Status", expanded=True):
    if data is None:
        st.error("Data: missing")
    else:
        st.success("Data: loaded")
        st.write(f"Rows: {len(data):,}")

    def status(name, obj, key=None):
        if obj is not None:
            try:
                cls = obj.__class__.__name__
            except Exception:
                cls = 'estimator'
            st.write(f"{name}: âœ… ({cls})")
        elif key and saved_label_arrays.get(key) is not None:
            st.write(f"{name}: âš ï¸ saved labels found (mapping available)")
        else:
            st.write(f"{name}: âŒ missing")

    status('K-Means', kmeans_model, 'kmeans')
    status('Agglomerative', agg_model, 'agg')
    status('Agglomerative (NN mapping)', agg_nn, 'agg')
    status('Mini-Batch KMeans', mb_kmeans_model, 'mb_kmeans')
    status('DBSCAN', dbscan_model, 'dbscan')
    status('Scaler', scaler)

# Page
st.title("ðŸŽ® Ronin Wallet Clustering Dashboard")
page = st.selectbox("Page", ["Dashboard", "Cluster Explorer", "User Predictor", "Business Intelligence"])

if page == "User Predictor":
    st.header("Predict User Segment")
    transaction_count = st.number_input("Transaction Count", min_value=0, value=150)
    days_active = st.number_input("Days Active", min_value=0, value=20)
    days_since_last = st.number_input("Days Since Last Transaction", min_value=0, value=2)
    value_sent = st.number_input("Total Value Sent (RON)", min_value=0.0, value=100.0)
    avg_value = st.number_input("Average Value per Transaction (RON)", min_value=0.0, value=1.0)
    unique_addresses = st.number_input("Unique Addresses", min_value=0, value=10)
    avg_gas = st.number_input("Average Gas Used", min_value=0, value=100000)
    wallet_age = st.number_input("Wallet Age (Days)", min_value=0, value=30)

    if st.button("ðŸ”® Predict Segment"):
        if scaler is None or kmeans_model is None:
            st.error("Models not loaded. Prediction is unavailable.")
        else:
            arr = np.array([[transaction_count, days_active, days_since_last, value_sent, avg_value, unique_addresses, avg_gas, wallet_age]])
            scaled = scaler.transform(arr)
            k_pred = int(kmeans_model.predict(scaled)[0])
            st.metric("K-Means", f"Cluster {k_pred}")

            # Agglomerative via NN mapping
            if agg_nn is not None and saved_label_arrays.get('agg') is not None:
                try:
                    idx = agg_nn.kneighbors(scaled, return_distance=False)[0][0]
                    agg_pred = int(saved_label_arrays['agg'][idx])
                    st.metric("Agglomerative", f"Cluster {agg_pred}")
                except Exception as e:
                    st.error(f"Agglomerative mapping failed: {e}")
            elif agg_model is not None:
                st.warning("Agglomerative: model present but cannot predict out-of-sample; provide agg_nn + agg_labels to enable mapping.")

            if mb_kmeans_model is not None:
                mb_pred = int(mb_kmeans_model.predict(scaled)[0])
                st.metric("Mini-Batch K-Means", f"Cluster {mb_pred}")

            if dbscan_model is not None:
                db_pred = dbscan_model.fit_predict(scaled)[0]
                st.metric("DBSCAN", "Outlier" if db_pred == -1 else f"Cluster {int(db_pred)}")

            # Show similar users from K-Means cluster (if available)
            if 'cluster' in data.columns:
                similar = data[data['cluster'] == k_pred]
                st.write(f"Users in this K-Means segment: {len(similar):,}")

# Export example (simple): export whales if column exists
if page == "Business Intelligence":
    st.header("Whale Monitoring")
    if 'value_sent' in data.columns:
        whale_threshold = st.slider("Whale Threshold (RON)", 1000, 10000, 5000, 500)
        whales = data[data['value_sent'] >= whale_threshold]
        st.metric("Total Whales", f"{len(whales):,}")
        if st.button("Export Whale List"):
            csv = whales.to_csv(index=False)
            st.download_button(label="Download Whales CSV", data=csv, file_name="whales.csv", mime="text/csv")

st.markdown("---")
st.write("Built from the notebook; for a dedicated app, save a cleaned `app.py` and run `streamlit run app.py`")
```
